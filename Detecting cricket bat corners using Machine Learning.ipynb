```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

# Applying Canny edge detection

```python
edges = cv2.Canny(gray, threshold1=50, threshold2=150)
```

# Finding the contours

```python
contours, _ = cv2.findContours(edges.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
for contour in contours:
    epsilon = 0.04 * cv2.arcLength(contour, True)
    approx = cv2.approxPolyDP(contour, epsilon, True)
    
    # Check if it's a quadrilateral (four corners)
    if len(approx) == 4:
        cv2.drawContours(image, [approx], -1, (0, 255, 0), 2)  # Draw the contour on the original image
        break  # Assuming you found the cricket bat, exit the loop

# Display the image with the detected cricket bat
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

# The 'approx' variable now contains the four corners of the cricket bat
print("Corners:", approx)
```
